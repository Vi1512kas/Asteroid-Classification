| Parameter                          | Description                                                                               |
|------------------------------------|-------------------------------------------------------------------------------------------|
| Absolute Magnitude (H)             | Intrinsic brightness of the asteroid, indicative of its size and potential impact energy. |
| Est Dia in KM (min) / (max)        | Estimated diameter range of the asteroid. Larger diameters indicate greater impact potential.|
| Epoch Date Close Approach          | Date and time of the asteroid's closest approach to Earth.                                 |
| Relative Velocity (km per sec)     | Speed of the asteroid relative to Earth. Faster speeds increase the impact potential.      |
| Miles per hour                     | Speed of the asteroid relative to Earth (in miles per hour).                                |
| Miss Dist. (kilometers)            | Closest distance between the asteroid's trajectory and Earth's orbit during close approach. |
| Hazardous                          | Indicates whether the asteroid is classified as "hazardous" (potential impact risk).        |
| Orbit Uncertainty                  | Level of uncertainty associated with the asteroid's predicted orbit.                        |
| Minimum Orbit Intersection (MOID)  | Minimum distance between the asteroid's orbit and Earth's orbit.                            |


```python
sns.pairplot(data,hue='Hazardous')
plt.legend(data['Hazardous'])
```

```python
def fun(data): 
    if data:
        return 1
    else:
        return 0
```

```python
data['Hazardous']=data['Hazardous'].apply(fun)
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer


cfobj = ColumnTransformer([
    ('obj1',StandardScaler(),[3,6])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
print(f"X_train shape :{X_train.shape}")
print(f"X_test shape :{X_test.shape}")
print(f"y_train shape :{y_train.shape}")
print(f"y_test shape :{y_test.shape}")
```

```python
model1 = LogisticRegression()
model1.fit(X_train,y_train)
y_pred = model1.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
model4 = LogisticRegression(penalty='l1',multi_class='ovr',solver='liblinear')
```

```python
model4.fit(X_train,y_train)
```

```python
y_pred2 = model4.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
pipe = Pipeline(
    steps=[
        ('step1',cfobj),
        # ('step2',DecisionTreeClassifier(criterion='entropy',max_depth=20,max_features='sqrt',min_samples_leaf=4,min_samples_split=10))
        ('step2',LogisticRegression(penalty='l1',multi_class='ovr',solver='liblinear'))
    ]
)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred4 = pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
import pickle
pickle.dump(pipe,open('pipeline.pkl','wb'))
```
